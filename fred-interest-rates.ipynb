```python
import os
from openbb import obb
#keys
obb_api_key = os.environ['OPENBB_API_KEY']
fred_api_key = os.environ['FRED_API_KEY']
obb.account.login(pat=obb_api_key)

obb.user.preferences.output_type = "dataframe"
```

```python

series_ids = ['DPRIME', 'DFF', 'DTB3', 'DTB6', 'DTB1YR','DGS1','DGS3', 'DGS5', 'DGS7','DGS10']
# FRED macro data
df_macro = obb.economy.fred_series(series_ids,frequency = 'm',start_date='1970-01-01')


df_macro.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
fig,axs = plt.subplots(3,1,figsize = (25,10),sharex=True)
sns.lineplot(df_macro[['DPRIME','DFF']],ax=axs[0])
sns.lineplot(df_macro[[ 'DTB3', 'DTB6', 'DTB1YR','DGS1']],ax=axs[1])
sns.lineplot(df_macro[['DGS3', 'DGS5', 'DGS7','DGS10']],ax=axs[2])
plt.suptitle('U.S. Federal Reserve Interest Rates over Time',fontsize=24)
plt.show()
```
